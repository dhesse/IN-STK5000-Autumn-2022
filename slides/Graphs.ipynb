```python
import matplotlib.pyplot as plt
import numpy
```

```python
with plt.rc_context({'font.size': 14}):
    plt.plot(numpy.linspace(0,100,100), label="Linear Growth")
    plt.plot(numpy.linspace(0,100, 100)**0.9, label="Imperfect Scaling")
    plt.legend()
    plt.xlabel('business size (arbitrary units)')
    plt.ylabel('business outcome (arbitrary units)')
plt.savefig("img/business-scaling.png", dpi=150, bbox_inches='tight')
```

```python
with plt.rc_context({'font.size': 14}):
    plt.plot(numpy.linspace(0,100,100), label="Linear Growth")
    plt.plot(numpy.linspace(0,100,100)**0.9, label="Imperfect Scaling")
    plt.plot(numpy.linspace(0,100,100)**1.2, label="Digital Business Model")
    plt.legend()
    plt.xlabel('business size (arbitrary units)')
    plt.ylabel('business outcome (arbitrary units)')
plt.savefig("img/digital-scaling.png", dpi=150)
```
